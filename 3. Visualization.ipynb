```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

# Plot the heatmap

```python
corr = df.corr()
```

```python

sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
plt.show()
```

# Plot the Boxplots for outliers

```python
# Create a box plot
plt.boxplot(df)

# Add title and labels as needed
plt.title('Box Plot')
plt.ylabel('Values')

# Show plot
plt.show()
```

```python
# Create a box plot
plt.boxplot(df.values)

# Setting x-tick labels
plt.xticks([1, 2, 3, 4], df.columns)
plt.ylabel('Values')

# Add title
plt.title('Box Plot')

# Show plot
plt.show()
```

```python
# Reshaping the DataFrame
df_melted = df.melt(var_name='Columns', value_name='Values')

# Create a box plot
sns.boxplot(x='Columns', y='Values', data=df_melted)

# Add title
plt.title('Box Plot')

# Show plot
plt.show()
```

### Distribution to see the outliers

```python
sns.histplot(data = df, x = 'newspaper', bins = 25 )
plt.show()
```

## Plot vs each feature one by one

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

# Pairplot - All Relationships between features

```python
# Relationships between features
sns.pairplot(df)
```

```python
# Relationships between features
sns.pairplot(df,diag_kind='kde')
```

```python
from pandas.plotting import scatter_matrix

# Plot scatter matrix
scatter_matrix(df, alpha=0.8, figsize=(10, 10), diagonal='kde')
plt.show()
```
