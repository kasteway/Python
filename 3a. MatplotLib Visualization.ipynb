### Customizing the line Charts:
    1. Line Style --- > linestyle or ls = '-. ' '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

## plt.plot(x, y, color = 'DeepSkyBlue', linestyle = '-.', linewidth = 10)

```python
x = [1,3,5,7,9]
y = [2,4,3,5,6]
```

```python
plt.plot(x, y, color = 'DeepSkyBlue', linestyle = '-.', linewidth = 10)
plt.title('Main TITLE', fontsize = 20, loc = 'left', color = 'b')
plt.xlabel('X-AXIS LABEL')
plt.ylabel('y-AXIS LABEL')
plt.grid();
```

## SubPlots to have Multiple Graphs in One Image
 
### subplot() requires 3 inputs:
    1. Number of Rows
    2. Number of Columns
    3. Index --> where the subplot is displayed starting at 1 not 0


### plt.subplot(nrows, ncols, index)

```python
plt.subplot(3,1,1)
plt.plot(x,y, color = 'r')

plt.subplot(3,1,2)
plt.plot(x,y, color = 'b')

plt.subplot(3,1,3)
plt.plot(x,y, color = 'k');
```

```python
plt.subplot(1,3,1)
plt.plot(x,y, color = 'r')

plt.subplot(1,3,2)
plt.plot(x,y, color = 'b')

plt.subplot(1,3,3)
plt.plot(x,y, color = 'k');
```
