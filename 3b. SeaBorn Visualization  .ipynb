```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (20, 10)
```

```python
df = pd.read_csv('../Pandas/hepatitis.csv')
df.head()
```

```python
#plt.figure(figsize=(8, 6), dpi=80)

plt.hist(df['sgot'], bins = 10, density=True);
```

    sns.histplot(column_name, bins = integer Number , stat= density, kde = True)

```python
sns.histplot(df['sgot'], bins = 10, stat = 'density', kde = True);
```

## Plots to visulaize Relationships

    1. Pair Plots
    2. Joint Plots

```python
sns.pairplot(df);
```

```python
sns.set_style('darkgrid')

sns.jointplot(x = 'age', y= 'target', kind = 'reg', data=df);
```

## HeatMaps

```python
avg = df.groupby(by = ['target']).mean()
```

```python
sns.heatmap(data = avg, cmap = 'Blues', annot = True);
```

## FacetGrid

```python
g = sns.FacetGrid(df, col = 'target')

g.map(sns.histplot, 'age');
```
